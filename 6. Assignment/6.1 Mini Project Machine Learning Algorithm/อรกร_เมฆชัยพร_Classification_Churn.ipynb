## แบ่งข้อมูล

```python
y = df_final['Churn_No Churn']
X = df_final.drop('Churn_No Churn', axis=1)
```

```python
y
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size = 0.3, random_state=555
)
```

```python
X_train
```

```python
y_train.value_counts()
```

```python
X_test
```

```python
y_test.value_counts()
```

## สร้าง Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier()
```

```python
clf.fit(X_train, y_train)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, \
precision_score, recall_score, classification_report
```

```python
y_pred = clf.predict(X_test)
```

```python
accuracy_score(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```

```python
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree
```

```python
X.columns
```

```python
classes = ('No Churn', 'Churn')
```

```python
plt.figure(figsize=(15, 15))
plot_tree(clf, feature_names=X.columns,
          class_names=classes, filled=True, rounded=True, max_depth=2)
plt.show()
```

```python
clf.feature_importances_
```

```python
Variable_importance = pd.DataFrame(dict(Feature=X.columns,
                  Value=clf.feature_importances_))\
  .sort_values(by='Value', ascending=False)
```

```python
Variable_importance
```

```python
from sklearn.metrics import RocCurveDisplay, roc_auc_score
```

```python
RocCurveDisplay.from_estimator(clf, X_test, y_test)
plt.show()
```

```python
roc_auc_score(y_test, clf.predict_proba(X_test)[:,1])
```

## สร้าง Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
```

# สรุป
เนื่องจากข้อมูลเป็น imbalance dataset เลย ใช้ Area Under Curve (AUC) ในการวัดผล ส่วนค่าการวัดผลอื่น ๆ ใน Confusion matrix สามารถดูได้ใน code ด้านบน

1. **ใช้ข้อมูลทั้งหมดเข้าโมเดล โดยปรับสเกลแล้ว:**
* Decision Tree ได้ AUC อยู่ที่ 0.86
* Random Forest (ใช้ estimatiors = 94) ได้ AUC อยู่ที่ 0.92


2. **ใช้ข้อมูลทั้งหมด ยกเว้นคอลัมน์ state, area code เข้าโมเดล โดยปรับสเกลแล้ว** (เพราะดูจาก feature_importances_ แล้วคอลัมน์ state, area codeไม่ค่อยมีผล) :
* Decision Tree ได้ AUC อยู่ที่  0.85
* Random Forest (ใช้ estimatiors = 94) ได้ AUC อยู่ที่ 0.93



จะเห็นว่า Random Forest ที่ใช้ข้อมูลทั้งหมด ยกเว้นคอลัมน์ state, area code เข้าโมเดล โดยปรับสเกลแล้ว ได้ค่า AUC มากที่สุด คือ **0.93**

หากอยากให้ model มีประสิทธิภาพมากกว่านี้น่าจะต้องลองเพิ่มข้อมูล และปรับ parameter ต่าง ๆ ดู
