# Ensemble Learning in Machine Learning

Ensemble Learning เป็นเทคนิคที่มีประสิทธิภาพที่เกี่ยวข้องกับการรวมตัวแบบจำนวนมากเพื่อสร้างโมเดลทำนายที่แม่นยำและทนทานมากขึ้น โดยการนำความสามารถของโมเดลต่าง ๆ มาใช้ร่วมกันและปรับปรุงจุดอ่อนของแต่ละโมเดล ทำให้ Ensemble Learning สามารถเพิ่มประสิทธิภาพในการทำนายได้อย่างมีนัยสำคัญ

## 1. Bagging (Bootstrap Aggregating):

Bagging เป็นกระบวนการฝึกโมเดลหลายรูปของโมเดลเดียวกันบนชุดข้อมูลการฝึกแตกต่างกันและรวมคำทำนายของพวกเขาเพื่อทำนายสุดท้าย ตัวอย่างที่น่าสนใจของ Bagging คือ อัลกอริทึม Random Forest

**Example:**
สมมติว่าคุณกำลังพยายามทำนายว่าลูกค้าจะซื้อผลิตภัณฑ์หรือไม่โดยอิงความเคลื่อนไหวของเว็บเบราว์ซิ่ง อายุ และพฤติกรรมการซื้อก่อนหน้านี้ คุณสร้างโมเดลอิสระที่เป็นตัวเลือกสามแบบ:
- Model A: Decision Tree
- Model B: Logistic Regression
- Model C: Support Vector Machine

สำหรับ Bagging คุณจะฝึกโมเดลแต่ละรูปบนชุดข้อมูลการฝึกที่แตกต่างกัน ในการทำนายคุณสามารถใช้การโหวตเพื่อรวมคำทำนายของสามโมเดลเหล่านี้ เช่น ถ้าโมเดล A ทำนาย "ซื้อ" โมเดล B ทำนาย "ไม่ซื้อ" และโมเดล C ทำนาย "ซื้อ" คำทำนายสุดท้ายอาจเป็น "ซื้อ" จากการโหวตส่วนใหญ่

## 2. Boosting:

Boosting เน้นการปรับปรุงความแม่นยำของโมเดลที่อ่อนแอโดยการฝึกพวกเขาตามลำดับ โมเดลที่ตามมาจะเน้นการแก้ไขข้อผิดพลาดของโมเดลที่พบไปก่อนหน้านี้

การทำนายสุดท้ายคือการผสมผสานคำทำนายของโมเดลแต่ละตัวโดยใช้ค่าน้ำหนัก

**Example:**
สมมติว่าคุณกำลังสร้างโมเดลเพื่อทำนายว่าผู้ป่วยมีโรคหรือไม่โดยอิงตัวชี้วัดสุขภาพต่าง ๆ คุณเริ่มโดยการฝึกโมเดล A แล้วตามด้วยการฝึกโมเดล B ด้วยข้อมูลที่โมเดล A ทำนายผิด และในลำดับต่อมาฝึกโมเดล C ด้วยข้อมูลที่โมเดล A และโมเดล B ทำนายผิด การทำนายสุดท้ายในกระบวนการ Boosting มักจะเป็นผลลัพธ์ที่ได้จากการรวมน้ำหนักของการทำนายของโมเดลทั้งสามรายการ

## 3. Stacking:

Stacking เป็นเทคนิคขั้นสูงของ Ensemble Learning ซึ่งนำเข้ามาใช้โมเดลเมต้าในการรวมการทำนายของโมเดลฐานหลาย ๆ รายการเข้าด้วยกัน โมเดลฐานจะทำนายผลบนข้อมูลนำเข้าและโมเดลเมต้าจะเรียนรู้วิธีกำหนดน้ำหนักและรวมการทำนายเหล่านี้ให้เหมาะสม

**Example:**
สมมติว่าคุณกำลังพัฒนาระบบแนะนำสินค้าสำหรับเว็บค้าออนไลน์ คุณสร้างโมเดลฐานสามรายการ:
- Model A: Decision Tree ที่ใช้ประวัติการเรียกดู
- Model B: Neural Network ที่ใช้ข้อมูลทางสถิติของผู้ใช้
- Model C: Gradient Boosting ที่ใช้ประวัติการซื้อและรีวิว

จากนั้นก็สร้างโมเดลเมต้า เช่น Random Forest Regression เพื่อรวมการทำนายของโมเดลฐานเหล่านี้ โมเดลเมต้าจะกำหนดน้ำหนักที่เหมาะสมให้กับการทำนายของแต่ละโมเดลฐาน ทำให้ได้ผลลัพธ์การแนะนำสินค้าสุดท้ายสำหรับผู้ใช้


# Recap of Decision Trees:

Decision Trees เป็นอัลกอริทึมที่ได้รับความนิยมสำหรับการเรียนรู้ของเครื่องใช้ทั้งในงาน classification และ prediction โดยสร้างโมเดลที่เปรียบเสมือนต้นไม้ที่มีการตัดสินใจและผลที่เป็นไปได้ทั้งหมด โหนดภายในแต่ละโหนดแทนคุณสมบัติหรือแอตทริบิวต์ แต่ละกิ่งแทนกฎการตัดสินใจและโหนดใบแทนผลลัพธ์

**Code Exercise: Building a Decision Tree Classifier**

ใช้ชุดข้อมูล Iris สร้าง Decision Tree โดยใช้ไลบรารี scikit-learn ของ Python

- Import ไลบรารีที่จำเป็น

```python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

- โหลด Iris dataset

```python
iris = load_iris()
iris
```

```python
X = iris.data
X
```

```python
y = iris.target
y
```

```python
len(y)
```

- แบ่งชุดข้อมูลเป็นชุดฝึกและชุดทดสอบ

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size = 0.3, random_state=555
)
```

- สร้าง Decision Tree

```python
clf = DecisionTreeClassifier()
```

- เรียนรู้ข้อมูล

```python
clf.fit(X_train, y_train)
```

- ทำนายผลด้วยข้อมูลทดสอบ

```python
y_pred = clf.predict(X_test)
accuracy_score(y_test, y_pred)
```

- แสดง Tree

```python
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree
```

```python
plot_tree(clf)
plt.show()
```

```python
print(export_text(clf, feature_names=iris.feature_names))
```

```python
plt.figure(figsize=(15, 15))
plot_tree(clf, feature_names=iris.feature_names,
          class_names=iris.target_names, filled=True, rounded=True)
plt.show()
```

# Random Forest Algorithm

The Random Forest algorithm is an ensemble learning technique that constructs multiple decision trees and aggregates their predictions. It introduces randomness in both data and feature selection during tree construction, leading to improved generalization and reduced overfitting. The final prediction is determined through a majority vote (for classification) or an average (for regression) of individual tree predictions.

# สร้าง DataFrame

- ใช้ dataset จากไฟล์ `winequality-red (INTERM DS).csv`
- อัพโหลดขึ้น sample_data
- เรียกใช้ `pandas`

- อ่านไฟล์ `winequality-red (INTERM DS).csv` จาก sample_data โดยคลิ๊กขวาที่ชื่อไฟล์ แล้วเลือก `Copy path`

```python
data = pd.read_csv('/content/sample_data/winequality-red (INTERM DS).csv')
data
```

```python
data['quality'].value_counts()
```

- ในที่นี้ต้องการทำนาย `quality`

```python
X = data.drop(['quality'], axis=1)
y = data['quality']
```

- แบ่งข้อมูลออกเป็นชุดการเรียนรู้ และชุดทดสอบ โดยใช้ `train_test_split`

- แบ่งข้อมูลสำหรับเรียน 80% สำหรับทดสอบ 20%
- ผลลัพธ์ที่ได้จะอยู่ในรูปของ
> `X_train, X_test, y_train, y_test`

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, train_size=0.8, random_state=555
)
```

```python
print(len(X_train), len(X_test))
```

# สร้าง RandomForest Classifier Model

- เรียกใช้งาน `RandomForestClassifier` library

```python
from sklearn.ensemble import RandomForestClassifier
```

- สร้าง model

```python
rf = RandomForestClassifier(n_estimators=82, random_state=555)
```

- เรียนรู้จากชุดข้อมูล

```python
rf.fit(X_train, y_train)
```

- ดู tree ย่อยที่อยู่ในชุด estimators

```python
rf.estimators_
```

```python
tree = rf.estimators_[0]
tree
```

- เรียกใช้เครื่องมือเพื่อช่วยแสดง tree

```python
from sklearn.tree import export_text
```

- แสดง tree แบบข้อความ

```python
print(export_text(tree, feature_names=list(X.columns)))
```

- ในกรณีนี้ class อยู่ในรูปแบบของตัวเลข
  - 0 - Bad Wine
  - 1 - Good Wine

```python
tree.classes_
```

- จึงสร้างเป็น tuple ของ class name ตาม label ที่มี

```python
wine_classes = ('Bad Wine', 'Good Wine')
```

- แสดง tree แบบกราฟ

```python
plt.figure(figsize=(100, 100))
plot_tree(tree, feature_names=X.columns,
          class_names=wine_classes, filled=True, rounded=True,
          max_depth=3)
plt.show()
```

- ใช้ model ที่ผ่านการเรียนรู้แล้วในการทำนาย

```python
predictions = rf.predict(X_test)
predictions
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cross_val_score(rf, X, y, cv=5).mean()
```

```python
scores = []
for n in range(50, 101):
  rf = RandomForestClassifier(n_estimators=n, random_state=555)
  scores.append(cross_val_score(rf, X, y, cv=5).mean())
max_score = max(scores)
max_index = scores.index(max_score)+50
print(max_index, max_score)
```
